```python

# 📌 Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 📌 Definir ruta del dataset limpio
data_path = "../data/final/sri_autos_final.csv"
```

```python
# 📌 Cargar el dataset
df = pd.read_csv(data_path, delimiter=';', encoding="latin-1", low_memory=False)
```

```python
# 📌 Mostrar información general del dataset
print("📌 Información del dataset:")
print(df.info())
```

```python

# 📌 Descripción estadística de las variables numéricas
print("\n📌 Estadísticas descriptivas:")
print(df.describe())
```

```python
# 📌 Histograma de "avaluo"
plt.figure(figsize=(10, 5))
sns.histplot(df["avaluo"], bins=30, kde=True, color="blue")
plt.xlabel("Avaluo ($)")
plt.ylabel("Cantidad de Autos")
plt.title("Distribución de Avalúos de Vehículos en Ecuador")
plt.show()
```

```python
# 📌 Histograma de "cilindraje"
plt.figure(figsize=(10, 5))
sns.histplot(df["cilindraje"], bins=30, kde=True, color="red")
plt.xlabel("Cilindraje")
plt.ylabel("Cantidad de Autos")
plt.title("Distribución del Cilindraje de Vehículos")
plt.show()
```

```python

# 📌 Relación entre "avaluo" y "year_modelo"
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["year_modelo"], y=df["avaluo"], alpha=0.5)
plt.xlabel("Año del Modelo")
plt.ylabel("Avaluo ($)")
plt.title("Relación entre Año del Modelo y Avalúo")
plt.show()
```

```python
# 📌 Relación entre "cilindraje" y "avaluo"
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["cilindraje"], y=df["avaluo"], alpha=0.5)
plt.xlabel("Cilindraje")
plt.ylabel("Avaluo ($)")
plt.title("Relación entre Cilindraje y Avalúo")
plt.show()
```
